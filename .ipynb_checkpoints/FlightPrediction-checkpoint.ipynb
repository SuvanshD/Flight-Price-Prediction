## Flight Price Prediction

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = pd.concat([train_df,test_df])
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
## Feature Engineering Processabs
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', inplace=True, axis=1)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
```

```python
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Dept_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min']=final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_min']=final_df['Dept_min'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops'].isnull().sum()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4, 'NaN':1})
```

```python
final_df['Total_Stops']
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df.drop('Route', axis=1, inplace=True )
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df.info()
```

```python
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['Duration_hour']=='5m']
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
```

```python
final_df['Duration_min']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Duration_min'].unique()
```

```python
final_df['Duration_min']=final_df['Duration_min'].apply(
    lambda x: '0' if x== 'nan' or x is np.nan else x)
```

```python
final_df['Duration_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Total_Duration_min'] = final_df['Duration_min'].astype(int) + final_df['Duration_hour'].astype(int) * 60
```
